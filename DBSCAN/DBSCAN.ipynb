```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

#Lets make dataset of 300 samples random with center 4 meaning the data can be grouped into 4 category
#random state= 42 meaning after random selection the data is supposed to be same even after shuffling when chosen randomly
X,_=make_blobs(n_samples=300,centers=4,random_state=42)

#Creating 20 random outliner data with 2 meaning (X,y)coordinates ranging from -15 to 15
outliners=np.random.uniform(low=-15,high=15,size=(20,2))

#inserting those 20 outliners to the actual 300 data in a stack
X=np.vstack([X,outliners])

#the data to be considered neigbour and not an outliner 1.5 is the maximum distance
#The data to be considered a category of group it must consist minimum 5 data
dbscan=DBSCAN(eps=1.5,min_samples=5)

#fit : To understand and extract information from data
#predict : To identify the category of the data so suppose data 1 as category 1, data 2 as category 0 ... basically labeling each data based on information extracted
labels=dbscan.fit_predict(X)

plt.figure(figsize=(8,6))#canvas of 8 inch width and 6 inch height
plt.scatter(X[:,0],X[:,1],c=labels) # the 1st column data i.e x and 2nd column data i.e y coordinate and c for colors
plt.title('DBSCAN Clusters') #Title
plt.show() #Display

print('Outliers found:',list(labels).count(-1))#Converting labels into list and counting the labels with value -1
print('Clusters found:',len(set(labels))-1) #Converting labels to set to remove duplicate and counting the remaining value and -1 to not count outliners

```
