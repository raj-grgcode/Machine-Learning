```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

#Creating a fake dataset.. So make_blobs is used to make fake dataset.
#Currently it consists of 300 samples and centers 4 and randomstate 42
from sklearn.datasets import make_blobs
X,_ = make_blobs(n_samples=300,centers=4,random_state=42)

#Using Elbow Method. We created a empty array inertia and range from 1 to 10
inertia=[]
K_range=range(1,11)

#So it takes k from 1 to 10 and for each K it checks Kmeans as k increases inertia decreases... 
#After a certain point even after k increases inertia(Grouping of data) doesnt really change much
#We take that point as a centre...
for k in K_range:
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X) #finds the paatterns in data alone, runs algorithm find centroids and assign clusters
    inertia.append(kmeans.inertia_)#basically adding total distance between all the points in each K so suppose 100 people in k=1 00 of them are in 1 grp k=2 80 in grp 1 20 in 2 and so on until its very less even after increasing k


plt.plot(K_range, inertia, marker='o')
plt.xlabel('Number of Clusters K')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
#As we know 4 is the right value so we directly took 4 as a value 
kmeans=KMeans(n_clusters=4,random_state=42)
#lets categorize value using fit
kmeans.fit(X)
labels=kmeans.labels_
#canvas of 8 height and 6 width inch 
plt.figure(figsize=(8, 6))

#So earlier we created 300 random data (x,y) and stored in X
#now X[:,0] Selects all 300 data from 1st column and X[:,1] from 2nd column
#c=labels gives color
plt.scatter(X[:, 0], X[:, 1], c=labels)

#Same concept as above but for centers only
plt.scatter(kmeans.cluster_centers_[:, 0], 
            kmeans.cluster_centers_[:, 1], 
            marker='*', s=50, c='red', label='Centroids')
plt.title('K-Means Clusters')
plt.legend()
plt.show()
```
