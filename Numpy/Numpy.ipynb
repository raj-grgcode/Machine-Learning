```python
#Numpy
import numpy as np
data=np.array([
 #10 students 5 subjects
#Math,Physics,Chemistry,Biology,English
    [85, 90, 78, 88, 92],
    [76, 81, 85, 79, 88],
    [90, 95, 89, 93, 91],
    [68, 72, 70, 75, 71],
    [88, 86, 90, 85, 89],
    [92, 88, 91, 90, 93],
    [74, 78, 76, 80, 79],
    [81, 85, 83, 87, 84],
    [95, 92, 96, 94, 97],
    [70, 73, 71, 74, 72]
])
print(data)
```

```python
print("***Creation Methods in Numpy***")
print("zeros")
print(np.zeros((3,3)))
print("Ones")
print(np.ones((2,4)))
print("empty")
print(np.empty((2,4)))
print("arange")
print(np.arange(0,20,2))
print("linspace")
print(np.linspace(0,100,11)) #11 evenly spaced numbers from 0-100
print("eye")
print(np.eye(4))
print("uniform random")
print(np.random.rand(3,3))
print("random integer")
print(np.random.randint(50,100,(5,3)))
      
```

```python
print("***Array Properties***")
print("shape")
print(data.shape)
print("dimension")
print(data.ndim)
print("data type and item size")
print(data.dtype)
print(data.itemsize)
```

```python
print("***Indexing & Slicing***")
print("1st student marks in maths",data[0,0])
print("1st student all marks",data[0])
print("all math scores",data[:,0])
print("all first three students, first 2 subjects")
print(data[0:3, 0:2])
print("student who scored more than 85 in maths",data[data[:,0]>85])
print("student who scored more than 85 in maths and physics"),data[(data[:,0]>85) & (data[:,1]>85)]
```

```python
print("***Reshaping***")
print("Reshaped to 5*10")
print(data.reshape(5,10))
print("Flattened to 1D")
print(data.flatten())
flat=data.flatten()
print("Flatten to 1D upto 20 elements")
print(flat[:20])
print("Add dimension")
expanded=np.expand_dims(data,axis=0)
print(expanded.shape)
print("Transpose")
print(data.T)
```
