```python
print("***Creation Methods in Numpy***")
print("zeros")
print(np.zeros((3,3)))
print("Ones")
print(np.ones((2,4)))
print("empty")
print(np.empty((2,4)))
print("arange")
print(np.arange(0,20,2))
print("linspace")
print(np.linspace(0,100,11)) #11 evenly spaced numbers from 0-100
print("eye")
print(np.eye(4))
print("uniform random")
print(np.random.rand(3,3))
print("random integer")
print(np.random.randint(50,100,(5,3)))
      
```
